```python
import pandas as pd
import numpy as np
import re
import nltk
import spacy
import string
pd.options.mode.chained_assignment = None
```

```python
full_df = pd.read_csv("twcs.csv")
full_df.head()
```

```python
df=full_df[['text']]
```

```python
df.head()
```

```python
df['text']=df['text'].astype(str)
df.head()
```

```python
df['text_lower']=df['text'].str.lower()
df.head()
```

```python
#drop last created column
df.drop(['text_lower'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
#remove punctuatuion 
punct_to_remove = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans(" "," ", punct_to_remove))
df['text_to_punct']=df['text_lower'].apply(lambda text:remove_punctuation(text))
df.head()
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
df['text_to_stop']=df['text_to_punct'].apply(lambda text: remove_stopwords(text))
df.head()
```

```python
from collections import Counter 
cnt = Counter()
for text in df['text_to_stop'].values:
    for word in text.split():
        cnt[word]+=1
        
    
cnt.most_common(10)
```

```python
FREQWORDS = [w for (w, wc) in cnt.most_common(10)]
def remove_freqwords(text):
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])
df['text_to_freqwords'] = df['text_to_stop'].apply(lambda text: remove_freqwords(text))
df.head()
```

```python
FREQWORDS
```

```python
df.drop(['text_to_punct','text_to_stop','text_lower'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#remove rare words
rare_word = 10
RARE_WORDS = [w for (w, wc) in cnt.most_common()[:-rare_word-1:-1]]
def remove_rarewords(text):
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])
df['text_to_rareword']= df['text_to_freqwords'].apply(lambda text: remove_freqwords(text))
df.head()
```

```python
RARE_WORDS
```

```python
#stemming
from nltk.stem.porter import PorterStemmer
```

```python
df.drop(['text_to_freqwords','text_to_rareword'],axis = 1, inplace = True)
```

```python
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
df["text_stemmed"] = df['text'].apply(lambda text: stem_words(text))
```

```python
df.head()
```

```python
#lemmetization
from nltk.stem import WordNetLemmatizer
lemmetizer = WordNetLemmatizer()
def lemmetize_word(text):
    return " ".join([lemmetizer.lemmatize(word) for word in text.split()])
df['lemmetized_word'] = df['text'].apply(lambda text: lemmetize_word(text))
df.head()
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
#remove urls
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'',text)
```

```python
#Remove html tages
def remove_htlm(text):
    html_patternv = re.compile('<.*?>')
    return html_pattern.sub(r'',text)
```

```python
lemmetizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmetize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmetizer.lemmatize(word, wordnet_map.get(pos[0],wordnet.NOUN)) for word, pos in pos_tagged_text])
df['text_lemmetized']= df['text'].apply(lambda text: lemmetize_words(text))
df.head()
```
